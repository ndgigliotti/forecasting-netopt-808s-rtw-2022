```python
import pandas as pd
from pulp import *

scrap_inventory = pd.read_csv(r'data\Challenge2_total_tires_public.csv')
vehicles = pd.read_csv(r'data\Challenge2_vehicles_public.csv')
destination_distances = pd.read_csv(r'data\Challenge2_distances_public.csv')
destination_capacities = pd.read_csv(r'data\Challenge2_capacity_public.csv')
destination_extra_ghg = pd.read_csv(r'data\Challenge2_second_leg_GHG_public.csv')
```

```python
destination_capacities.rename(columns={"max_tire_capacity":"max_tire_capacity_destination"}, inplace=True)
vehicles.rename(columns={"max_tire_capacity":"max_tire_capacity_vehicle"}, inplace=True)

print("scrap_inventory")
display(scrap_inventory)

print('vehicles')
display(vehicles)

print('destination_distances')
display(destination_distances)

print('destination_capacities')
display(destination_capacities)

print('destination_extra_ghg')
display(destination_extra_ghg)
```

```python
#all_combos = pd.merge(scrap_inventory, vehicles, on="dc_name")


all_combos = pd.merge(vehicles, destination_distances, on='dc_name')
all_combos = pd.merge(all_combos, destination_capacities, left_on='destination', right_on='location')
all_combos = pd.merge(all_combos, destination_extra_ghg, left_on='location', right_on='location')
all_combos = pd.merge(all_combos, destination_extra_ghg, left_on='dc_name', right_on='location')

all_combos.rename(columns={"base_ghg_x":"base_ghg_destination_disposal"}, inplace=True)
all_combos.rename(columns={"additional_ghg_per_tire_x":"additional_ghg_per_tire_destination_disposal"}, inplace=True)

all_combos.rename(columns={"base_ghg_y":"base_ghg_origin_disposal_BAD"}, inplace=True)
all_combos.rename(columns={"additional_ghg_per_tire_y":"additional_ghg_per_tire_origin_disposal_BAD"}, inplace=True)

all_combos = all_combos.drop(columns=['location_y', 'location_x'])

all_combos['full_route'] = all_combos['dc_name'] + "_" + all_combos['vehicle_name'] + "_" + all_combos['destination']
```
